```python
import sys
from matplotlib import pyplot
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# define cnn model
def define_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(200, 200, 3)))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(1, activation='sigmoid'))
    # compile model
    opt = SGD(learning_rate=0.0001, momentum=0.9)
    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# plot diagnostic learning curves
# plot diagnostic learning curves
def summarize_diagnostics(history):
    # plot loss
    pyplot.figure(figsize=(10, 5))
    pyplot.subplot(211)
    pyplot.title('Cross Entropy Loss')
    pyplot.plot(history.history['loss'], color='blue', label='train')
    pyplot.plot(history.history['val_loss'], color='orange', label='test')
    pyplot.legend(['train', 'test'], loc='upper right')

    # plot accuracy
    pyplot.subplot(212)
    pyplot.title('Classification Accuracy')
    pyplot.plot(history.history['accuracy'], color='blue', label='train')
    pyplot.plot(history.history['val_accuracy'], color='orange', label='test')
    pyplot.legend(['train', 'test'], loc='upper right')

    # display the plot
    pyplot.tight_layout()  # Adjust layout for better display
    pyplot.show()          # Show plot in the output terminal


# run the test harness for evaluating a model
def run_test_harness():
    # define model
    model = define_model()
    # create data generator
    datagen = ImageDataGenerator(rescale=1.0/255.0)
    # prepare iterators
    train_it = datagen.flow_from_directory('dataset_dogs_vs_cats/train/',
                                           class_mode='binary', batch_size=64, target_size=(200, 200))
    test_it = datagen.flow_from_directory('dataset_dogs_vs_cats/test/',
                                          class_mode='binary', batch_size=64, target_size=(200, 200))
    # fit model
    history = model.fit(train_it, steps_per_epoch=len(train_it),
                        validation_data=test_it, validation_steps=len(test_it), epochs=20, verbose=1)
    # evaluate model
    _, acc = model.evaluate(test_it, steps=len(test_it), verbose=1)
    print('> %.3f' % (acc * 100.0))
    # learning curves
    summarize_diagnostics(history)

# entry point, run the test harness
run_test_harness()
```

```python
import os
import time
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import VGG16, VGG19
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_dir = "dataset_dogs_vs_cats"  # Root directory for your dataset
train_dir = os.path.join(data_dir, "train")  # Path to training data
test_dir = os.path.join(data_dir, "test")    # Path to testing data

# Prepare the dataset
data_gen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Training and validation generators
train_gen = data_gen.flow_from_directory(
    train_dir,
    target_size=(200, 200),
    batch_size=64,
    class_mode='binary',
    subset='training')

val_gen = data_gen.flow_from_directory(
    train_dir,
    target_size=(200, 200),
    batch_size=64,
    class_mode='binary',
    subset='validation')

# Testing generator
test_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    test_dir,
    target_size=(200, 200),
    batch_size=64,
    class_mode='binary')
```

```python
results = []

def log_results(model_name, training_time, train_loss, train_acc, test_acc, num_params):
    results.append({
        "Model": model_name,
        "Training Time (s)": training_time,
        "Training Loss": train_loss,
        "Training Accuracy": train_acc,
        "Testing Accuracy": test_acc,
        "Number of Parameters": num_params
    })
```

```python

def vgg_3_blocks():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(200, 200, 3)),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    return model
```

```python
def transfer_learning_model(base_model, fine_tune=False):
    base_model.trainable = fine_tune
    model = models.Sequential([
        base_model,
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    return model
```

```python
def train_and_evaluate(model, model_name, augmented_data_gen=None):
    model.compile(optimizer=optimizers.Adam(), loss='binary_crossentropy', metrics=['accuracy'])

    start_time = time.time()
    if augmented_data_gen:
        history = model.fit(augmented_data_gen, epochs=10, validation_data=val_gen)
    else:
        history = model.fit(train_gen, epochs=10, validation_data=val_gen)
    training_time = time.time() - start_time

    train_loss, train_acc = history.history['loss'][-1], history.history['accuracy'][-1]
    test_acc = history.history['val_accuracy'][-1]
    num_params = model.count_params()

    log_results(model_name, training_time, train_loss, train_acc, test_acc, num_params)
```

```python
img_size = (200, 200)
model = vgg_1_block()
train_and_evaluate(model, "VGG (1 block)")

# 2. VGG (3 blocks)
model = vgg_3_blocks()
train_and_evaluate(model, "VGG (3 blocks)")

# 3. Transfer learning - fine-tune all layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(*img_size, 3))
model = transfer_learning_model(base_model, fine_tune=True)
train_and_evaluate(model, "Transfer Learning (VGG16, fine-tune all layers)")

# 4. Transfer learning - fine-tune only MLP layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(*img_size, 3))
model = transfer_learning_model(base_model, fine_tune=False)
train_and_evaluate(model, "Transfer Learning (VGG16, fine-tune MLP layers only)")

# Print results
import pandas as pd
results_df = pd.DataFrame(results)
print(results_df)

# Save results to a CSV
results_df.to_csv("model_comparison_results.csv", index=False)
```

```python
import tensorflow as tf
from tensorflow.keras.callbacks import TensorBoard
import datetime
import os
import numpy as np

# TensorBoard log directory
log_dir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1, write_images=True)

# Custom callback for logging additional metrics and images
class CustomTensorBoardCallback(tf.keras.callbacks.Callback):
    def __init__(self, log_dir, test_gen):
        super().__init__()
        self.writer = tf.summary.create_file_writer(log_dir)
        self.test_gen = test_gen
        self.step = 0

    def on_train_batch_end(self, batch, logs=None):
        # Log training loss and accuracy
        with self.writer.as_default():
            tf.summary.scalar("Training Loss", logs["loss"], step=self.step)
            tf.summary.scalar("Training Accuracy", logs["accuracy"], step=self.step)
            self.writer.flush()
        self.step += 1

    def on_test_batch_end(self, batch, logs=None):
        # Log testing accuracy
        with self.writer.as_default():
            tf.summary.scalar("Testing Accuracy", logs["accuracy"], step=self.step)
            self.writer.flush()

    def on_epoch_end(self, epoch, logs=None):
        # Log images and their predictions
        images, labels = next(iter(self.test_gen))  # Get a batch of test data
        predictions = self.model.predict(images)

        with self.writer.as_default():
            for i in range(len(images)):
                tf.summary.image(
                    f"Test Image {i}",
                    np.expand_dims(images[i], axis=0),
                    step=epoch,
                    description=f"Prediction: {predictions[i][0]:.2f}, Label: {labels[i]}"
                )
            self.writer.flush()

# Initialize the custom callback
custom_callback = CustomTensorBoardCallback(log_dir, test_gen)
```
