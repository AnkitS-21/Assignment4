```python
from os import listdir
from os.path import isfile, join
from numpy import asarray, save
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Define location of dataset
folder = 'train/'  # Make sure this path is correct

# Initialize lists for photos and labels
photos, labels = list(), list()

# Check if folder exists and iterate over files in the directory
try:
    files = [file for file in listdir(folder) if isfile(join(folder, file))]
    for file in files:
        # Check for image file extensions
        if file.endswith('.jpg') or file.endswith('.png') or file.endswith('.jpeg'):
            # Determine class based on filename
            output = 1.0 if file.startswith('dog') else 0.0
            
            # Load image
            photo = load_img(join(folder, file), target_size=(200, 200))
            
            # Convert to numpy array
            photo = img_to_array(photo)
            
            # Append to lists
            photos.append(photo)
            labels.append(output)
    
    # Convert lists to numpy arrays
    photos = asarray(photos)
    labels = asarray(labels)
    
    # Output the shape of the arrays
    print(photos.shape, labels.shape)
    
    # Save the reshaped arrays
    save('dogs_vs_cats_photos.npy', photos)
    save('dogs_vs_cats_labels.npy', labels)
    
except FileNotFoundError:
    print(f"Directory '{folder}' not found. Please check the path.")
```
